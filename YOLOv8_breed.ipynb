```python
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the trained YOLOv8 models
model1 = YOLO("/content/drive/MyDrive/CMPE258_Project/breed/results/best.pt")
model2 = YOLO("/content/drive/MyDrive/CMPE258_Project/emotion/results/best.pt")

# Load the image
image = cv2.imread("/content/drive/MyDrive/CMPE258_Project/breed/images/pred/Beagle.jpg")

# Get class labels for both models
class_labels_model1 = model1.names  # Class labels for model1
class_labels_model2 = model2.names  # Class labels for model2

# Load the image
image = cv2.imread("/content/drive/MyDrive/CMPE258_Project/breed/images/pred/Beagle.jpg")

# Predict with both models
results1 = model1.predict(source=image, save=False, conf=0.5)
results2 = model2.predict(source=image, save=False, conf=0.5)

# Function to combine predictions with separate label sets
from collections import defaultdict

def combine_predictions(results1, results2):
    combined_results = defaultdict(list)

    # Process results from model1
    for box in results1[0].boxes.data.tolist():  # Assuming batch size = 1
        x1, y1, x2, y2, conf, cls = box[:6]
        combined_results[("model1", int(cls))].append((x1, y1, x2, y2, conf))

    # Process results from model2
    for box in results2[0].boxes.data.tolist():  # Assuming batch size = 1
        x1, y1, x2, y2, conf, cls = box[:6]
        combined_results[("model2", int(cls))].append((x1, y1, x2, y2, conf))

    return combined_results

# Combine predictions
combined_results = combine_predictions(results1, results2)

# Visualize combined predictions
def visualize_predictions(image, combined_results, labels1, labels2):
    # Define colors for each model
    color_model1 = (0, 0, 0)  # Red for model1
    color_model2 = (255, 255, 255)  # Green for model2

    image_copy = image.copy()
    for (model_id, cls), boxes in combined_results.items():
        for box in boxes:
            x1, y1, x2, y2, conf = map(int, box)

            # Determine the label set and color based on the model
            if model_id == "model1":
                label_name = labels1.get(cls, f"Class {cls}")
                color = color_model1
            else:
                label_name = labels2.get(cls, f"Class {cls}")
                color = color_model2

            # Draw the bounding box and label
            cv2.rectangle(image_copy, (x1, y1), (x2, y2), color, 2)
            label = f"{label_name}"
            cv2.putText(image_copy, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    return image_copy

# Visualize and display the combined results
output_image = visualize_predictions(image, combined_results, class_labels_model1, class_labels_model2)
cv2_imshow(output_image)  # Display the image in Colab
```
